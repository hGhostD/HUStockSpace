```python
df_sorted = df.sort_values(by='成交额', ascending=False)
```

```python
d = df_sorted.loc[:, ['代码', '名称', '最新价', '涨跌幅', '成交额', '振幅', '量比', '换手率']]
d['成交额'] = d['成交额'] / 1e8  # 除以 1 亿
d['成交额'] = d['成交额'].map(lambda x: f'{x:.2f}亿')  # 格式化为字符串
d = d.reset_index(drop=True)
d.index += 1
d.head(50)
```

```python
d_filtered = d.loc[d['代码'] == '000002']
d_filtered
```

```python
d_filtered['量比'].values[0]
```

量比小于0.5 属于明显缩量。缩量能创新高，说明主力高度控盘，而且能排除主力出货的可能，如果正好**处于<span style="color: #FF0022;">上升</span>通道**，则吃肉的概率极大。
如果股票涨停量比小于 1，说明还有较大的上涨空间，次日再次涨停的概率极大。
如果量比大于1.5，而且突破某个重要主力位，是不可多得的买入位。

```python
def checkVolumeRatio(item):
    print(item)
    if isinstance(item['量比'], pd.Series):
        ratio = float(item['量比'].iloc[0])
    else:
        ratio = float(item['量比'])
    print('量比:', ratio)
    if ratio < 0.5:
        print('属于明显缩量，主力高度控盘能创新高。')
    elif ratio < 1:
        print('还有较大的上涨空间，次日再次涨停的概率极大。')
    elif ratio > 1.5:
        print('如果突破了主力位，就是不可多得的买入位置。')
    else:
        print('还不符合特殊的量比数据')
```
