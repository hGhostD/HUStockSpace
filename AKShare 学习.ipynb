量比小于0.5 属于明显缩量。缩量能创新高，说明主力高度控盘，而且能排除主力出货的可能，如果正好**处于<span style="color: #FF0022;">上升</span>通道**，则吃肉的概率极大。
如果股票涨停量比小于 1，说明还有较大的上涨空间，次日再次涨停的概率极大。
如果量比大于1.5，而且突破某个重要主力位，是不可多得的买入位。

```python
def checkVolumeRatio(item):
    print(item)
    if isinstance(item['量比'], pd.Series):
        ratio = float(item['量比'].iloc[0])
    else:
        ratio = float(item['量比'])
    print('量比:', ratio)
    if ratio < 0.5:
        print('属于明显缩量，主力高度控盘能创新高。')
    elif ratio < 1:
        print('还有较大的上涨空间，次日再次涨停的概率极大。')
    elif ratio > 1.5:
        print('如果突破了主力位，就是不可多得的买入位置。')
    else:
        print('还不符合特殊的量比数据')
```

```python
stock_share_hold_change_szse_df = ak.stock_share_hold_change_szse(symbol="002771")
q = stock_share_hold_change_szse_df['变动日期'].head(50)
q
```

```python
# 将数据转换为 DataFrame


# 将数据转换为日期格式
q = pd.to_datetime(q)

# 提取年份和月日
years = q.dt.year
day_month = q.dt.strftime('%m-%d')

# 创建 DataFrame
df = pd.DataFrame({'Year': years, 'Day_Month': day_month})

# 创建一个新的列，将每个日期转换为一个数值，以便绘图
df['Day_Month_Numeric'] = pd.to_datetime(df['Day_Month'], format='%m-%d').dt.dayofyear

# 绘制折线图
plt.plot(df['Year'], df['Day_Month_Numeric'], marker='o')

# 添加标题和标签
plt.title('日期折线图 (年份 vs. 月日)')
plt.xlabel('年份')
plt.ylabel('月日 (数值)')

# 显示网格
plt.grid()
plt.xticks(df['Year'].unique())  # 只显示唯一的年份
plt.tight_layout()  # 自适应布局
plt.show()
```
